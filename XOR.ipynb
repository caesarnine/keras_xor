```python
# import all the required libraries
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import Adam, Adadelta
import numpy as np
```

```python
# set a random seed and define the inputs and outputs for XOR
np.random.seed(5)
inputs = np.array([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
outputs = np.array([[0.],[1.],[1.],[0.]])
```

```python
# build the network with two input nodes, three hidden nodes, and one output node
model = Sequential()
model.add(Dense(output_dim=3,input_dim=2))
model.add(Activation('relu',))
model.add(Dense(output_dim=1))
model.add(Activation('relu'))
```

```python
# compile the model, using AdaDelta as the optimizer
model.compile(loss='mse',optimizer=Adadelta())
```

```python
# fit the model for 500 epochs
model.fit(inputs,outputs,nb_epoch=500,batch_size=4)
```

```python
# make a predicton on the inputs
model.predict(inputs)
```
